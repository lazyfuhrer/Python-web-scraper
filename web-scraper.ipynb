```python
#  Importing libraries

import requests
from bs4 import BeautifulSoup
import pandas as pd
url = "https://fmoviesto.cc/movie"
```

```python
#  Making a GET request to the website

res = requests.get(url)
```

```python
#  Saving the response as a html file

with open('web-page.html', 'w') as f:
    f.write(res.text)
```

```python
#  Creating a doc object of BeautifulSoup

doc = BeautifulSoup(res.text, 'html.parser')    
```

```python
#  Getting all the titles using BeautifulSoup libraries

title_tags = doc.find_all('h2', {'class': 'film-name'})
title_tags_list = []
for tag in title_tags:
    title_tags_list.append(tag.text.strip())
```

```python
#  Getting all the years and duration value

desc_tags = doc.find_all('span', {'class': 'fdi-item'})
year_tags = ['2021']
duration_tags = []

for i in range(1, len(desc_tags)):
    if i%2 != 0:
        duration_tags.append(desc_tags[i].text)
    else:
        year_tags.append(desc_tags[i].text)    
```

```python
#  Getting all the movie links from elements

links = doc.find_all('a')

list = []
for i in links:
    list.append(i.get('href'))
    
lst = []
for item in list:
    if not item in lst:
        lst.append(item)    
        
lst = lst[75:107]      

movie_links = []
for item in lst:
    new_link = "https://fmoviesto.cc"+item
    movie_links.append(new_link)       
```

```python
# Making a dictionary to parse this data into pandas object named topics_df 

topics_dict = {
    'Movie Name': title_tags_list,
    'Release Year': year_tags,
    'Duration': duration_tags,
    'Link' : movie_links
}
```

```python
topics_df = pd.DataFrame(topics_dict)  
```
